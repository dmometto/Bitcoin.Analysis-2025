```python
import pandas as pd

bitcoin_df = pd.read_csv('datasets-bitcoin.csv', sep=',')
bitcoin_df['Timestamp'] = pd.to_datetime(bitcoin_df['Timestamp'], unit='s')

bitcoin_df['Year'] = bitcoin_df['Timestamp'].dt.year
bitcoin_df['Month'] = bitcoin_df['Timestamp'].dt.month_name()
bitcoin_df['Day'] = bitcoin_df['Timestamp'].dt.day

bitcoin_df_2025 = bitcoin_df[bitcoin_df['Year'] == 2025].copy()

dates_df = bitcoin_df_2025[
    (bitcoin_df_2025['Timestamp'].dt.day == 1) | 
    (bitcoin_df_2025['Timestamp'].dt.day == 15) | 
    (bitcoin_df_2025['Timestamp'].dt.is_month_end)
]

bitcoin_filtrado_df = dates_df.drop(columns=['Timestamp']).drop_duplicates(subset=['Year', 'Month', 'Day']).reset_index(drop=True)

bitcoin_filtrado_df['Variacao_Pct'] = bitcoin_filtrado_df['Close'].pct_change() * 100 
bitcoin_filtrado_df['Variacao_Pct'] = bitcoin_filtrado_df['Variacao_Pct'].round(2).fillna(0)

bitcoin_filtrado_df.to_csv('bitcoin_2025.csv', index=False)
```
