```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

df = pd.read_csv('bitcoin_2025.csv')

ordem_meses = [
    'January', 'February', 'March', 'April', 'May', 'June', 
    'July', 'August', 'September', 'October', 'November', 'December'
]

df['Month'] = pd.Categorical(df['Month'], categories = ordem_meses, ordered = True)

plt.figure(figsize=(12, 6))

grafico = sns.barplot(
    data = df,
    x = 'Month',
    y = 'Variacao_Pct',
    estimator = 'mean',
    palette = 'RdYlGn', 
    hue = 'Month',  
    legend = False
 )

plt.title('Bitcoin 2025 : Variação Média Mensal (%)', fontsize = 15, fontweight = 'bold')
plt.ylabel('Variação Média (%)')
plt.xlabel('Mês')

plt.axhline(0, color='black', linewidth=0.8)

plt.show()
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

df2 = pd.read_csv('bitcoin_2025.csv')

df2['Amplitude'] = df2['High'] - df2['Low']

plt.figure(figsize=(12, 6))

grafico_vol = sns.barplot(
    data=df2, 
    x='Month', 
    y='Amplitude', 
    estimator='mean', 
    palette='Oranges',
    hue='Month',
    legend=False
)

plt.title('Volatilidade Média Mensal do Bitcoin (2025)', fontsize=15, fontweight='bold')
plt.ylabel('Amplitude Média (Diferença High - Low)')
plt.xlabel('Mês')

plt.show()
```
